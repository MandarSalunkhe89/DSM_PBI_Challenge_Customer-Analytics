```python
import pandas as pd
import random
import calendar

# Create a list of months for the year
months = [month for month in calendar.month_name if month]

# Generate random data for 1000 rows
data = []
for _ in range(1000):
    month = random.choice(months)
    year = random.randint(2020, 2023)
    monthly_sales = random.randint(50000, 150000)
    avg_transaction_value = random.uniform(20, 100)
    customer_retention_rate = random.uniform(0.4, 0.8)
    new_customers_acquired = random.randint(50, 200)

    data.append([month, year, monthly_sales, avg_transaction_value, customer_retention_rate, new_customers_acquired])

# Create a DataFrame
columns = ['Month', 'Year', 'Monthly Sales', 'Average Transaction Value', 'Customer Retention Rate', 'New Customers Acquired']
df = pd.DataFrame(data, columns=columns)

# Export the DataFrame to a CSV file
df.to_csv('retail_kpi_metrics.csv', index=False)
```
